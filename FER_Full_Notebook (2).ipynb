```python
# ============================================================
# 1) IMPORTS & CONSTANTS
# ============================================================
import tensorflow as tf
from tensorflow.keras import layers, models
import os
from PIL import Image
import io

DATA_ROOT = "/content/drive/MyDrive/FER_Project/data"
IMG_SIZE = (48, 48)
BATCH_SIZE = 32
EPOCHS = 12  # fewer epochs


# ============================================================
# 2) CLEAN DATASET (REMOVE CORRUPTED IMAGES)
# ============================================================
def clean_dataset(root):
    VALID_EXT = (".jpg", ".jpeg", ".png", ".bmp")
    removed = 0
    valid_files_count = 0

    print("Starting dataset cleaning...")

    for split in ["train", "val", "test"]:
        base = os.path.join(root, split)
        if not os.path.exists(base):
            print(f"Warning: {base} does not exist. Skipping split.")
            continue

        for emotion in os.listdir(base):
            folder = os.path.join(base, emotion)
            if not os.path.isdir(folder):
                continue

            for f in list(os.listdir(folder)): # Iterate over a copy as we might remove files
                path = os.path.join(folder, f)

                # Remove Mac hidden files
                if f.startswith("._"):
                    os.remove(path)
                    removed += 1
                    # print(f"Removed __MACOSX file: {path}")
                    continue

                # Remove bad formats or tiny files
                if not f.lower().endswith(VALID_EXT) or os.path.getsize(path) < 200:
                    os.remove(path)
                    removed += 1
                    # print(f"Removed invalid format or tiny file: {path}")
                    continue

                # Try loading the image with PIL and then with TensorFlow
                try:
                    with Image.open(path) as img_pil:
                        img_pil.verify() # Check for PIL-level corruption

                    # Try decoding with TensorFlow to catch specific TF decoding issues
                    img_bytes = tf.io.read_file(path)
                    _ = tf.io.decode_image(img_bytes, channels=1, expand_animations=False) # Try decoding as grayscale
                    valid_files_count += 1

                except (Image.UnidentifiedImageError, OSError, ValueError) as e:
                    os.remove(path)
                    removed += 1
                    print(f"Removed corrupted image (PIL/TF decode error): {path} - {e}")
                except tf.errors.InvalidArgumentError as e:
                    os.remove(path)
                    removed += 1
                    print(f"Removed corrupted image (TF InvalidArgumentError): {path} - {e}")
                except Exception as e:
                    os.remove(path)
                    removed += 1
                    print(f"Removed file due to unexpected error during validation: {path} - {e}")

    print(f"Finished dataset cleaning. Total files removed: {removed}. Total valid files remaining: {valid_files_count}")

clean_dataset(DATA_ROOT)


# ============================================================
# 3) LOAD DATASETS SAFELY
# ============================================================
print("Loading datasets...")
try:
    train_ds = tf.keras.preprocessing.image_dataset_from_directory(
        os.path.join(DATA_ROOT, "train"),
        label_mode="categorical",
        color_mode="grayscale",
        image_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        shuffle=True
    )

    val_ds = tf.keras.preprocessing.image_dataset_from_directory(
        os.path.join(DATA_ROOT, "val"),
        label_mode="categorical",
        color_mode="grayscale",
        image_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        shuffle=False
    )

    test_ds = tf.keras.preprocessing.image_dataset_from_directory(
        os.path.join(DATA_ROOT, "test"),
        label_mode="categorical",
        color_mode="grayscale",
        image_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        shuffle=False
    )

    num_classes = len(train_ds.class_names)
    print("Classes:", train_ds.class_names)

except Exception as e:
    print(f"Error loading datasets: {e}. Please ensure DATA_ROOT ('{DATA_ROOT}') is correctly populated with images.")
    raise # Re-raise the exception to stop execution


# ============================================================
# 4) PREFETCH FOR SPEED
# ============================================================
AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.prefetch(AUTOTUNE)
val_ds = val_ds.prefetch(AUTOTUNE)
test_ds = test_ds.prefetch(AUTOTUNE)


# ============================================================
# 5) BUILD A SIMPLE & STABLE CNN MODEL
# ============================================================
def build_cnn(num_classes):
    model = models.Sequential([
        layers.Input(shape=(48, 48, 1)),

        layers.Conv2D(32, (3,3), activation="relu"),
        layers.MaxPooling2D(),

        layers.Conv2D(64, (3,3), activation="relu"),
        layers.MaxPooling2D(),

        layers.Conv2D(128, (3,3), activation="relu"),
        layers.MaxPooling2D(),

        layers.Flatten(),
        layers.Dense(128, activation="relu"),
        layers.Dropout(0.4),

        layers.Dense(num_classes, activation="softmax")
    ])

    model.compile(
        optimizer="adam",
        loss="categorical_crossentropy",
        metrics=["accuracy"]
    )
    return model

model = build_cnn(num_classes)
model.summary()


# ============================================================
# 6) CALLBACKS (EARLY STOPPING, LR REDUCTION)
# ============================================================
callbacks = [
    tf.keras.callbacks.EarlyStopping(
        monitor="val_loss",
        patience=3,
        restore_best_weights=True
    ),
    tf.keras.callbacks.ReduceLROnPlateau(
        monitor="val_loss",
        factor=0.5,
        patience=2
    )
]


# ============================================================
# 7) TRAIN MODEL
# ============================================================
print("Starting model training...")
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS,
    callbacks=callbacks
)


# ============================================================
# 8) EVALUATE ON TEST SET
# ============================================================
print("Evaluating model on test set...")
test_loss, test_acc = model.evaluate(test_ds)
print(f"\nTest Accuracy: {test_acc:.4f}")
```
